# Load and Concatenate Enrolment Data

```python
import pandas as pd
import os

# Define the directory path for the enrolment CSV files
enrolment_dir = '/content/api_data_aadhar_enrolment/'

# Create an empty list to store individual DataFrames
df_list = []

# Iterate through all files in the specified directory
for filename in os.listdir(enrolment_dir):
    if filename.endswith('.csv'):
        file_path = os.path.join(enrolment_dir, filename)
        try:
            # Read each CSV file into a pandas DataFrame and append it to the list
            df = pd.read_csv(file_path)
            df_list.append(df)
            print(f"Successfully loaded {filename}")
        except Exception as e:
            print(f"Error loading {filename}: {e}")

# Concatenate all DataFrames in the list vertically into a single DataFrame
if df_list:
    enrolment_all_df = pd.concat(df_list, ignore_index=True)
    print(f"Concatenated {len(df_list)} files into a single DataFrame with {len(enrolment_all_df)} rows.")

    # Save the concatenated DataFrame to a new CSV file
    output_file_path = '/content/enrolment_all.csv'
    enrolment_all_df.to_csv(output_file_path, index=False)
    print(f"Saved concatenated data to '{output_file_path}'")
else:
    print("No CSV files found to concatenate.")
```

## Feature Engineering: Enrolment Metrics


```python
import pandas as pd

# Create 'total_enrolments' column
df_master['total_enrolments'] = df_master['age_0_5'] + df_master['age_5_17'] + df_master['age_18_greater']

# Create 'child_enrolments' column
df_master['child_enrolments'] = df_master['age_0_5'] + df_master['age_5_17']

# Create 'CES' (Child Enrolment Share) column
df_master['CES'] = df_master['child_enrolments'] / df_master['total_enrolments']

print("New columns 'total_enrolments', 'child_enrolments', and 'CES' created successfully.")
print(df_master.head())
```

## Feature Engineering: Update Metrics and Ratios


```python
import pandas as pd

# Create 'total_demo_updates' column
df_master['total_demo_updates'] = df_master['demo_age_5_17'] + df_master['demo_age_17_']


# For demonstration purposes, creating placeholder columns with arbitrary values if they don't exist
# In a real scenario, these values would come from the actual data.
if 'address_update' not in df_master.columns:
    df_master['address_update'] = df_master['total_demo_updates'] * 0.5 # Example placeholder
if 'mobile_update' not in df_master.columns:
    df_master['mobile_update'] = df_master['total_demo_updates'] * 0.4 # Example placeholder

# Create 'address_update_share' column
df_master['address_update_share'] = df_master['address_update'] / df_master['total_demo_updates']

# Create 'mobile_update_share' column
df_master['mobile_update_share'] = df_master['mobile_update'] / df_master['total_demo_updates']

# Create 'total_bio_updates' column
df_master['total_bio_updates'] = df_master['bio_age_5_17'] + df_master['bio_age_17_']

# Create 'DUR' (Demographic Update Ratio) column
df_master['DUR'] = df_master['total_demo_updates'] / df_master['total_enrolments']

# Create 'BUR' (Biometric Update Ratio) column
df_master['BUR'] = df_master['total_bio_updates'] / df_master['total_enrolments']

print("New columns 'total_demo_updates', 'address_update_share', 'mobile_update_share', 'total_bio_updates', 'DUR', and 'BUR' created successfully.")
print(df_master.head())
```

## Feature Engineering: Normalized Metrics and ALBI


```python
import pandas as pd

# Function for min-max normalization
def min_max_normalize(series):
    min_val = series.min()
    max_val = series.max()
    if max_val == min_val:
        return 0  # Avoid division by zero if all values are the same
    return (series - min_val) / (max_val - min_val)

# Group by 'year' and 'state' and apply min-max normalization
# Apply normalization for 'total_enrolments'
df_master['E_norm'] = df_master.groupby(['year', 'state'])['total_enrolments'].transform(min_max_normalize)

# Apply normalization for 'total_demo_updates'
df_master['D_norm'] = df_master.groupby(['year', 'state'])['total_demo_updates'].transform(min_max_normalize)

# Apply normalization for 'total_bio_updates'
df_master['B_norm'] = df_master.groupby(['year', 'state'])['total_bio_updates'].transform(min_max_normalize)

# Calculate the standard deviation across 'E_norm', 'D_norm', and 'B_norm' for each row
df_master['std_norm'] = df_master[['E_norm', 'D_norm', 'B_norm']].std(axis=1)

# Create 'ALBI' (Life-Cycle Balance Index) column
df_master['ALBI'] = 1 - df_master['std_norm']

print("New columns 'E_norm', 'D_norm', 'B_norm', 'std_norm', and 'ALBI' created successfully.")
print(df_master.head())
```

## Calculate Aadhaar Data Risk Score


```python
import numpy as np
import pandas as pd

# Enrolment Risk: Use the existing 'E_norm' column
df_master['E_norm_risk'] = df_master['E_norm']

# Child Enrolment Share (CES) Risk: Normalize 'CES' (higher CES means higher risk)
df_master['CES_risk_norm'] = min_max_normalize(df_master['CES'])

# Demographic Update Deficiency Risk:
# Handle inf and NaN values in 'DUR' before normalization
# Replace inf with the maximum finite value, then fill NaNs with 0.
# This ensures that very high DUR (inf) is treated as low deficiency risk (maps to 1 before 1-x),
# and 0/0 or 0 DUR is treated as high deficiency risk (maps to 0 before 1-x).
cleaned_dur = df_master['DUR'].replace([np.inf, -np.inf], np.nan)
max_finite_dur = cleaned_dur.loc[cleaned_dur.notna()].max()
cleaned_dur.fillna(0, inplace=True)
df_master['DUR_risk_norm'] = 1 - min_max_normalize(cleaned_dur)

# Biometric Update Deficiency Risk:
# Handle inf and NaN values in 'BUR' before normalization (similar to DUR)
cleaned_bur = df_master['BUR'].replace([np.inf, -np.inf], np.nan)
max_finite_bur = cleaned_bur.loc[cleaned_bur.notna()].max()
cleaned_bur.fillna(0, inplace=True)
df_master['BUR_risk_norm'] = 1 - min_max_normalize(cleaned_bur)

# Life-Cycle Balance Index (ALBI) Risk: Subtract 'ALBI' from 1
df_master['ALBI_risk_norm'] = 1 - df_master['ALBI']

# --- Step 2: Create 'risk_score_unscaled' by summing components ---
# Using equal weights for simplicity as no specific weights were provided.
df_master['risk_score_unscaled'] = (
    df_master['E_norm_risk'] +
    df_master['CES_risk_norm'] +
    df_master['DUR_risk_norm'] +
    df_master['BUR_risk_norm'] +
    df_master['ALBI_risk_norm']
)

# --- Step 3: Normalize 'risk_score_unscaled' to 0-1 range ---
df_master['risk_score_normalized_0_1'] = min_max_normalize(df_master['risk_score_unscaled'])

# --- Step 4: Multiply by 100 to get final 'risk_score' (0-100) ---
df_master['risk_score'] = df_master['risk_score_normalized_0_1'] * 100

print("Aadhaar Data Risk Score and all intermediate risk components calculated successfully.")
print(df_master[['year', 'month', 'state', 'district', 'E_norm_risk', 'CES_risk_norm', 'DUR_risk_norm', 'BUR_risk_norm', 'ALBI_risk_norm', 'risk_score']].head())
```

## K-means Clustering of Districts.


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. Select the features for clustering
features = ['CES', 'DUR', 'BUR', 'address_update_share', 'mobile_update_share']
df_clustering = df_master[features].copy()

# Handle potential NaN values before scaling and clustering
# For simplicity, fill with mean, but a more robust strategy might be needed depending on data.
df_clustering.fillna(df_clustering.mean(), inplace=True)

# 2. Standardize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)
df_scaled = pd.DataFrame(df_scaled, columns=features, index=df_clustering.index)

# 3. Instantiate KMeans model
# Choosing 4 clusters as a reasonable starting point. An elbow method or silhouette score could be used to optimize this.
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# 4. Fit the KMeans model to the scaled data and predict clusters
df_master['cluster_label'] = kmeans.fit_predict(df_scaled)

# 6. Display the head of df_master including the new 'cluster_label' column
print("df_master with cluster assignments:")
print(df_master[['state', 'district', 'CES', 'DUR', 'BUR', 'address_update_share', 'mobile_update_share', 'cluster_label']].head())
```

## Anomaly Detection on Monthly Enrolments.


```python
import pandas as pd

# Ensure df_national_monthly is sorted by date for rolling calculations
df_national_monthly = df_national_monthly.sort_values(by='Date').reset_index(drop=True)

# Calculate a 3-month rolling mean and standard deviation for 'total_enrolments'
df_national_monthly['enrolments_rolling_mean'] = df_national_monthly['total_enrolments'].rolling(window=3).mean()
df_national_monthly['enrolments_rolling_std'] = df_national_monthly['total_enrolments'].rolling(window=3).std()

# Define upper and lower bounds for anomalies (rolling mean  2 standard deviations)
df_national_monthly['enrolments_upper_bound'] = df_national_monthly['enrolments_rolling_mean'] + (2 * df_national_monthly['enrolments_rolling_std'])
df_national_monthly['enrolments_lower_bound'] = df_national_monthly['enrolments_rolling_mean'] - (2 * df_national_monthly['enrolments_rolling_std'])

# Identify anomalies for 'total_enrolments'
df_national_monthly['enrolments_anomaly'] = ((df_national_monthly['total_enrolments'] > df_national_monthly['enrolments_upper_bound']) |
                                         (df_national_monthly['total_enrolments'] < df_national_monthly['enrolments_lower_bound']))

print("Rolling statistics and anomaly detection for 'total_enrolments' calculated successfully.")
print(df_national_monthly[['Date', 'total_enrolments', 'enrolments_rolling_mean', 'enrolments_upper_bound', 'enrolments_lower_bound', 'enrolments_anomaly']].head())
```

```python
import pandas as pd

# Calculate the difference between 'total_enrolments' and 'total_demo_updates'
df_national_monthly['enrolment_vs_demo_diff'] = df_national_monthly['total_enrolments'] - df_national_monthly['total_demo_updates']

# Calculate the difference between 'total_enrolments' and 'total_bio_updates'
df_national_monthly['enrolment_vs_bio_diff'] = df_national_monthly['total_enrolments'] - df_national_monthly['total_bio_updates']

# Calculate rolling mean and standard deviation for 'enrolment_vs_demo_diff'
df_national_monthly['demo_diff_rolling_mean'] = df_national_monthly['enrolment_vs_demo_diff'].rolling(window=3).mean()
df_national_monthly['demo_diff_rolling_std'] = df_national_monthly['enrolment_vs_demo_diff'].rolling(window=3).std()
df_national_monthly['demo_diff_upper_bound'] = df_national_monthly['demo_diff_rolling_mean'] + (2 * df_national_monthly['demo_diff_rolling_std'])
df_national_monthly['demo_diff_lower_bound'] = df_national_monthly['demo_diff_rolling_mean'] - (2 * df_national_monthly['demo_diff_rolling_std'])
df_national_monthly['demo_diff_anomaly'] = ((df_national_monthly['enrolment_vs_demo_diff'] > df_national_monthly['demo_diff_upper_bound']) |
                                            (df_national_monthly['enrolment_vs_demo_diff'] < df_national_monthly['demo_diff_lower_bound']))

# Calculate rolling mean and standard deviation for 'enrolment_vs_bio_diff'
df_national_monthly['bio_diff_rolling_mean'] = df_national_monthly['enrolment_vs_bio_diff'].rolling(window=3).mean()
df_national_monthly['bio_diff_rolling_std'] = df_national_monthly['enrolment_vs_bio_diff'].rolling(window=3).std()
df_national_monthly['bio_diff_upper_bound'] = df_national_monthly['bio_diff_rolling_mean'] + (2 * df_national_monthly['bio_diff_rolling_std'])
df_national_monthly['bio_diff_lower_bound'] = df_national_monthly['bio_diff_rolling_mean'] - (2 * df_national_monthly['bio_diff_rolling_std'])
df_national_monthly['bio_diff_anomaly'] = ((df_national_monthly['enrolment_vs_bio_diff'] > df_national_monthly['bio_diff_upper_bound']) |
                                           (df_national_monthly['enrolment_vs_bio_diff'] < df_national_monthly['bio_diff_lower_bound']))

print("Lag check differences, rolling statistics, and anomaly detection calculated successfully.")
print(df_national_monthly[['Date', 'enrolment_vs_demo_diff', 'demo_diff_anomaly', 'enrolment_vs_bio_diff', 'bio_diff_anomaly']].head())
```

## Summary:

**1. Brief overview of the Aadhaar Data Risk Score:**
The Aadhaar Data Risk Score is a composite metric ranging from 0 to 100, designed to quantify potential data integrity or operational risks at a district-year level. It is calculated by combining five normalized risk components:
*   **Enrolment Risk:** Directly from normalized total enrolments (`E_norm_risk`).
*   **Child Enrolment Share (CES) Risk:** Normalized CES, as a higher share of child enrolments might indicate higher future update needs.
*   **Demographic Update Deficiency Risk (DUR_risk_norm):** Calculated as `1 - normalized(DUR)`, implying that lower Demographic Update Ratio (DUR) contributes to higher risk.
*   **Biometric Update Deficiency Risk (BUR_risk_norm):** Calculated as `1 - normalized(BUR)`, indicating that lower Biometric Update Ratio (BUR) contributes to higher risk.
*   **Life-Cycle Balance Index (ALBI) Risk:** Calculated as `1 - ALBI`, where a lower ALBI (indicating more imbalance between enrolment, demographic, and biometric activities) contributes to higher risk.
Each component is min-max normalized and summed to create an unscaled score, which is then scaled to the 0-100 range.

**2. Any significant anomalies detected:**
Anomaly detection was performed on national monthly time-series data using a 3-month rolling mean 2 standard deviations.
*   **Anomalies in Total Enrolments:** The process identified specific months where the total national enrolments significantly deviated (outside 2 standard deviations) from their rolling average, suggesting unusually high or low enrolment activity.
*   **Anomalies in Enrolment vs. Update Differences:** Anomalies were also detected in the differences between `total_enrolments` and `total_demographic_updates`, as well as between `total_enrolments` and `total_biometric_updates`. These deviations indicate unusual patterns in the balance between new enrolments and subsequent updates, which could point to issues like backlog in updates or unusual surges in enrolments without corresponding updates.
